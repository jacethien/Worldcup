```python
# Lần lượt tạo ra 4 biểu đồ bar chart với dữ liệu world_cups ở trên
b1 = px.bar(data_frame=worldcups,
            x='Year',
            y='Attendance',
            color='Year', text='Attendance', orientation='v')

b2 = px.bar(data_frame=worldcups,
            x='Year',
            y='QualifiedTeams',
            color='Year', text='QualifiedTeams', orientation='v')

b3 = px.bar(data_frame=worldcups,
            x='Year',
            y='GoalsScored',
            color='Year', text='GoalsScored', orientation='v')

b4 = px.bar(data_frame=worldcups,
            x='Year',
            y='MatchesPlayed',
            color='Year', text='MatchesPlayed', orientation='v')


# Tạo ra 1 subplots có 4 dòng (rows = 4) và 1 cột (cols = 1)
# vertical_spacing: Khoảng cách theo chiều dọc với các subplots khác. Ở đây mình để 0.05
# subplot_titles: List có chứa tên các title của các subplots
kq = make_subplots(rows=4,cols=1,vertical_spacing=0.05,
                  subplot_titles=['Số người tham gia từng năm','Số đội tham gia từng năm',
                                 'Số bàn thắng từng năm', 'Số trận đầu từng năm'])

# Tạo ra các trace tương ứng từ 4 biểu đồ đã vẽ
trace_1 = b1['data'][0]
trace_2 = b2['data'][0]
trace_3 = b3['data'][0]
trace_4 = b4['data'][0]

# Add lần lượt các trace vào fig với vị trí row và col tương ứng bạn muốn
kq.add_trace(trace_1,row=1,col=1)
kq.add_trace(trace_2,row=2,col=1)
kq.add_trace(trace_3,row=3,col=1)
kq.add_trace(trace_4,row=4,col=1)

# ta sử dụng câu lệnh trên để chuyển dữ liệu trục x từ dạng số sang dạng Category.
# Nếu để nguyên thì nó sẽ tiếp diễn kiểu 0 5 10 ... chứ không liên tục nữa
kq.update_xaxes(type='category')

# Thay đổi layout, tăng chiều cao lên 2000 và không show ra coloraxis
kq.update_layout(height=2000,coloraxis_showscale=False)

# Show biểu đồ ra
kq.show()
```
